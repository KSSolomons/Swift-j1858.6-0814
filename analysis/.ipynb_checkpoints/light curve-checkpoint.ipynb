```python
from astropy.table import Table

# Replace with the path to your file
lc_file_path = "../products/0865600201/pn/pn_src_lc.fits"

# Read the file by specifying the HDU (extension) name
lc_table = Table.read(lc_file_path, hdu="RATE")

# You can now access the columns like a dictionary
time = lc_table['TIME']
rate = lc_table['RATE']
rate_error = lc_table['ERROR']

# Print the table to see its contents
print(lc_table)
```

```python
#plotting the light curve
import matplotlib.pyplot as plt

plt.errorbar(time, rate, yerr=rate_error, fmt='o', ecolor='red', capsize=2)
plt.xlabel('Time (s)')
plt.ylabel('Count Rate (counts/s)')
plt.title('Light Curve')
plt.grid()
plt.show()
```
